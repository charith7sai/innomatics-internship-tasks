```python
# Problem: Check if username and password match "admin" and "1234"
username = "admin"  
password = "1234"   

if username == "admin" and password == "1234":
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]
pass_count = 0
fail_count = 0

for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Total Pass: {pass_count}")
print(f"Total Fail: {fail_count}")
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    msg_len = len(msg)
    if msg_len > 10:
        print(f"{msg_len}: {msg} (Long Message)")
    else:
        print(f"{msg_len}: {msg}")
```
