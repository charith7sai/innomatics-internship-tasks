```python
def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    while True:
        amount = int(input("Enter recharge amount: "))
        if amount >= 50 and amount in valid_plans:
            print("Recharge Successful")
            break
        else:
            print("Invalid Amount or Plan. Please try again.")

validate_recharge()
```

```python
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {"Laptops": 20, "Mice": 10, "Keyboards": 14, "Monitors": 25}
check_inventory(inventory)
```

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark
    average = total / len(marks)
    return "Pass" if average >= 50 else "Fail"

student_marks = [65, 70, 45, 80, 55]
print(process_result(student_marks))
```

```python
def calculate_fare(km, is_peak):
    fare = 50 + (km * 12)
    if is_peak:
        fare *= 1.25
    return fare

while True:
    dist = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower() == 'yes'
    print(f"Total Fare: ₹{calculate_fare(dist, peak)}")
    
    retry = input("Calculate again? (yes/no): ").lower()
    if retry != 'yes':
        break
```
