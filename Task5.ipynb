```python
def analyze_engagement(daily_likes):
    total_likes = 0
    # Using a loop for calculations as requested
    for likes in daily_likes:
        total_likes += likes
        
    print(f"Total Likes: {total_likes}")
    
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")

# Example usage
analyze_engagement([250, 300, 400, 100])
```

```python
def check_medicine_stock(stock_level):
    print(f"Medicine Stock: {stock_level}")
    
    if stock_level < 10:
        print("Status: Low Stock Alert")
    else:
        print("Status: Stock Sufficient")

# Example usage
check_medicine_stock(6)
```

```python
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0
    days = 0
    
    # Using a loop to calculate the sum for the average
    for rain in rainfall_data:
        total_rainfall += rain
        days += 1
        
    average_rainfall = total_rainfall / days
    print(f"Average Rainfall: {average_rainfall:.0f}")
    
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")

# Example usage
check_rainfall([65, 80, 70, 73], 70)
```

```python
def detect_duplicate_accounts(usernames):
    # A set removes duplicates. If lengths differ, duplicates exist.
    unique_accounts = set(usernames)
    
    if len(unique_accounts) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example usage
detect_duplicate_accounts(["user123", "ai_dev", "data_fan", "user123"])
```

```python
def check_appointment_eligibility(age):
    print(f"Patient Age: {age}")
    
    if age >= 18:
        print("Eligibility Status: Eligible")
    else:
        print("Eligibility Status: Not Eligible")

# Example usage
check_appointment_eligibility(21)
```

```python
def filter_premium_crops(crop_prices):
    premium_crops = []
    
    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)
            
    print(f"Premium Crops: {premium_crops}")

# Example usage
filter_premium_crops([1500, 2500, 1800, 3200, 1900])
```

```python
def check_application_health(error_count):
    print(f"Error Count: {error_count}")
    
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")

# Example usage
check_application_health(7)
```

```python
def validate_transaction(amount):
    daily_limit = 50000
    print(f"Transaction Amount: {amount}")
    
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")

# Example usage
validate_transaction(60000)
```

```python
def check_attendance_eligibility(attendance_record):
    days_present = 0
    total_days = len(attendance_record)
    
    # Using a loop to count present days
    for day in attendance_record:
        if day == 1:
            days_present += 1
            
    attendance_percentage = (days_present / total_days) * 100
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    
    if attendance_percentage >= 75.0:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")

# Example usage (8 days present out of 10)
check_attendance_eligibility([1, 1, 1, 1, 0, 1, 1, 1, 1, 0])
```
