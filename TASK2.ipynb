```python
sentence = "python is easy and python is powerful"
words = sentence.split()
unique_words = set(words)

print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

```python
numbers = [45, 22, 89, 10, 66]
max_val = numbers[0]
min_val = numbers[0]

for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}")
print(f"Min: {min_val}")
```

```python
prices = [450, 1200, 899, 1500, 300]
count = 0

for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

```python
attendance = ["P", "P", "A", "P", "P"]
present_days = attendance.count("P")
total_days = len(attendance)
percentage = (present_days / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

```python
phone_numbers = [9876543210, 9123456789, 9876543210]
unique_numbers = set(phone_numbers)

print(f"Unique phone numbers: {unique_numbers}")
```

```python
text = "python"
frequency = {}

for char in text:
    frequency[char] = frequency.get(char, 0) + 1

print(frequency)
```
